# Exploratory Data Analysis
## What-If Sports NIL Simulator

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
%matplotlib inline
```

```python
# Load sample data
df = pd.read_csv('../data/sample/sample_players.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Basic stats
df.describe()
```

```python
# NIL tier distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
df['nil_tier'].value_counts().sort_index().plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('NIL Tier Distribution')
axes[0].set_xlabel('Tier')

df['nil_valuation'].hist(bins=30, ax=axes[1], color='steelblue')
axes[1].set_title('NIL Valuation Distribution')
axes[1].set_xlabel('Valuation ($)')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
numeric_cols = ['ppg', 'apg', 'rpg', 'games_played', 'program_tier', 'nil_valuation', 'injury_flag']
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# NIL valuation over time for a few players
sample_players = df['player_id'].unique()[:5]
plt.figure(figsize=(12, 5))
for pid in sample_players:
    player = df[df['player_id'] == pid]
    plt.plot(player['snapshot_week'], player['nil_valuation'], marker='o', label=pid)
plt.xlabel('Week')
plt.ylabel('NIL Valuation ($)')
plt.title('NIL Valuation Trajectories')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# NIL by program tier
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='program_tier', y='nil_valuation', palette='viridis')
plt.title('NIL Valuation by Program Tier')
plt.xlabel('Program Tier')
plt.ylabel('NIL Valuation ($)')
plt.show()
```
